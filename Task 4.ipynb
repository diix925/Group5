```python
df = pd.read_csv('gas_prices.csv')

plt.figure(figsize=(8,5))

plt.title('Gas Prices over Time (in USD)' )

plt.plot(df.Year, df.USA, 'b.-')
plt.plot(df.Year, df.Canada,  'r.-')
plt.plot(df.Year,df['South Korea'], 'g.-')

print(df.Year[::3])

plt.xticks(df.Year[::3])

plt.xlabel('Year')
plt.ylabel('US Dollars')

plt.legend

plt.show()
```

```python
df = pd.read_csv("fifa_data.csv")
df.head(5)
```

```python
bins = [40,50,60,70,80,90,100]

plt.hist(df.Overall, bins=bins)

plt.xticks(bins)

plt.ylabel('Number of Players')
plt.xlabel('Skill Level')
plt.title('Distribution of players Skills FIFA 2018')

plt.show()
```

```python
left =df.loc[df['Preferred Foot'] == 'Left'].count()[0]
right =df.loc[df['Preferred Foot'] == 'Right'].count()[0]

labels = ['Left' ,'Right']
colors = ['#abcdef', '#aabbcc']

plt.pie([left, right], labels = labels, colors = colors, autopct='%.2f %%')

plt.title('Preferenc of FIFA Players')

plt.show()
```
